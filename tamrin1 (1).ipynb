# many to one with simple lstm

```python
# پانزده داده تولید کردم که هر کدوم 3 تا تایم استپ دارن و لیبل هم مجموع این 3 مقدار هست
# فعلا هم از حالت تک فیچره استفاده کردم 

X = np.array([x+1 for x in range(45)])
print(X)
```

```python
X = X.reshape(15,3,1)
print(X)
```

```python
Y = list()
for x in X:
    Y.append(x.sum())

Y = np.array(Y)
print(Y)
```

## solve it with pure lsm 

```python
model = Sequential()
model.add(LSTM(50, activation='relu', input_shape=(3, 1)))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')
```

```python
history = model.fit(X, Y, epochs=1000, validation_split=0.2, verbose=1)
```

```python
test_input = array([50,51,52]) # 50 + 51 + 52 = 153 
test_input = test_input.reshape((1, 3, 1))
test_output = model.predict(test_input, verbose=0)
print(test_output)
```

# I choose stack lstm 

```python
model = Sequential()
model.add(LSTM(200, activation='relu', return_sequences=True, input_shape=(3, 1)))
model.add(LSTM(100, activation='relu', return_sequences=True))
model.add(LSTM(50, activation='relu', return_sequences=True))
model.add(LSTM(25, activation='relu'))
model.add(Dense(20, activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')

history = model.fit(X, Y, epochs=1000, validation_split=0.2, verbose=1)
```

```python
test_output = model.predict(test_input, verbose=0)
print(test_output) # مقدار بهتری رو نسبت به مدل قبلی پیش بینی کرد 
```

```python
# پانزده دیتا رو تولید میکنیم و لیبل رو اینطوری میزنیم که به ازای هر ورودی ... مدل باید دو عدد بعدیشو در دستگاه اعداد صحیح به ما بده 
# یعنی اگر 4 رو بدیم باید 5 و 6 رو به عنوان خروجی به ما بده 

X = list()
Y = list()
X = [x+3 for x in range(-2, 43, 3)]

for i in X:
    output_vector = list()
    output_vector.append(i+1)
    output_vector.append(i+2)
    Y.append(output_vector)

print(X)
print(Y)
```

# solve it using stacking 

```python
from keras.layers import RepeatVector
from keras.layers import TimeDistributed

model = Sequential()

# encoder layer
model.add(LSTM(100, activation='relu', input_shape=(3, 1)))

# repeat vector
model.add(RepeatVector(3))

# decoder layer
model.add(LSTM(100, activation='relu', return_sequences=True))

model.add(TimeDistributed(Dense(1)))
model.compile(optimizer='adam', loss='mse')

print(model.summary())
```

```python
history = model.fit(X, Y, epochs=1000, validation_split=0.2, verbose=1, batch_size=3)
```

```python
test_input = array([300, 305, 310])
test_input = test_input.reshape((1, 3, 1))
test_output = model.predict(test_input, verbose=0)
print(test_output)
```
